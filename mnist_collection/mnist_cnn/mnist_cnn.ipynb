# MNIST Handwritten Digits Convolutional Neural Network 

```python
import numpy as np
import torch
from torchvision import datasets,transforms
from torch.utils.data.sampler import SubsetRandomSampler
```

```python
transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,),(0.5,))])

trainOnGPU = torch.cuda.is_available()

if trainOnGPU:
    print("CUDA available.. training on GPU")
else:
    print("CUDA not available.. training on CPU")
    
trainSet = datasets.MNIST("data",download = True, train=True, transform = transform)
testSet = datasets.MNIST("data",download = True, train = False,transform = transform)
```

```python
numOfWorkers = 0
batchSize = 20
valSize = 0.2

numOfTrain = len(trainSet)
indices = list(range(numOfTrain))
np.random.shuffle(indices)
split = int(np.floor(numOfTrain * valSize))
trainIdx,valIdx = indices[split:],indices[:split]
trainSampler = SubsetRandomSampler(trainIdx)
valSampler = SubsetRandomSampler(valIdx)

trainLoader = torch.utils.data.DataLoader(trainSet,batch_size = batchSize, sampler = trainSampler,num_workers = numOfWorkers)
valLoader = torch.utils.data.DataLoader(trainSet,batch_size = batchSize, sampler = valSampler, num_workers = numOfWorkers)
testLoader = torch.utils.data.DataLoader(testSet,batch_size = batchSize)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

it = iter(trainLoader)
images,labels = it.next()
images = images.numpy()
fig = plt.figure(figsize = (25,5))
for idx in np.arange(20):
    ax = fig.add_subplot(2,20/2,idx+1,xticks=[],yticks=[])
    ax.imshow(np.squeeze(images[idx]),cmap="gray")
    ax.set_title(str(labels[idx].item()))
```

```python
from torch import nn,optim

class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        # sees 28x28x1
        self.conv1 = nn.Conv2d(1,16,3,padding = 1)
        # sees 14x14x16
        self.conv2 = nn.Conv2d(16,32,3,padding = 1)
        # sees 7x7x32
        self.conv3 = nn.Conv2d(32,64,3,padding = 1)
        # sees 3x3x64
        self.pool = nn.MaxPool2d(2,2)
        self.fc1 = nn.Linear(64*3*3, 256)
        self.fc2 = nn.Linear(256,128)
        self.fc3 = nn.Linear(128,64)
        self.fc4 = nn.Linear(64,10)
        self.relu = nn.ReLU()
        self.softmax = nn.LogSoftmax(dim=1)
        self.dropout= nn.Dropout(0.2)
    def forward(self,x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = self.pool(self.relu(self.conv3(x)))
        x = x.view(-1,64*3*3)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.relu(self.fc3(x))
        x = self.dropout(x)
        x = self.fc4(x)
        return self.softmax(x)

model = Classifier()
print(model)
```

```python
import torch.optim as optim
criterion = nn.NLLLoss()
optimizer = optim.SGD(model.parameters(),lr=0.01)
```

```python
numOfEpochs = 20

for epoch in range(numOfEpochs):
    trainLoss = 0.0
    valLoss = 0.0
    testLoss = 0.0
    accuracy = 0.0
    model.train()
    for imgs,lbls in trainLoader:
        if trainOnGPU:
            imgs,lbls = imgs.cuda(),lbls.cuda()
        
        optimizer.zero_grad()
        output = model(imgs)
        loss = criterion(output,lbls)
        loss.backward()
        optimizer.step()
        trainLoss += loss.item()*imgs.size(0)
    
    model.eval()
    with torch.no_grad():
        for imgs,lbls in valLoader:
            if trainOnGPU:
                imgs,lbls = imgs.cuda(),lbls.cuda()
            output = model(imgs)
            loss = criterion(output,lbls)
            valLoss += loss.item()*imgs.size(0)
        for imgs,lbls in testLoader:
            if trainOnGPU:
                imgs,lbls = imgs.cuda(),lbls.cuda()
            log_ps = model(imgs)
            testLoss += criterion(output,lbls)
            ps = torch.exp(log_ps)
            top_p,top_class = ps.topk(1,dim=1)
            equals = top_class == lbls.view(*top_class.shape)
            accuracy += torch.mean(equals.type(torch.FloatTensor))
    trainLoss = trainLoss / len(trainLoader.sampler)
    valLoss = valLoss / len(valLoader.sampler)
    accuracy = accuracy / len(testLoader)*100
    print("Epoch: {}\tTrain Loss: {:.6f}\tValdation Loss: {:.6f}\tAccuracy: {:.2f}".format(epoch + 1,trainLoss,valLoss,accuracy))
    
```

```python
model.eval()
it = iter(testLoader)
imgs,lbls = it.next()
output = model(imgs)

_,preds = torch.max(output,1)
fig = plt.figure(figsize=(25,4))
for idx in np.arange(20):
    ax = fig.add_subplot(2,20/2,idx+1,xticks=[],yticks=[])
    ax.imshow(np.squeeze(imgs[idx]),cmap="gray")
    ax.set_title("{} ({})".format(str(preds[idx].item()),str(lbls[idx].item())),
                color = ("green" if preds[idx] == lbls[idx] else "red"))
```
