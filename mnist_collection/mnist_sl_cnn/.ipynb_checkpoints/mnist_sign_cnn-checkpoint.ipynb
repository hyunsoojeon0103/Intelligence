# MNIST Sign Language Using Convolutional Neural Network

```python
import numpy as np
from torchvision import transforms,datasets
import torch
import pandas as pd
from PIL import Image
```

```python
# define custom dataset
class DriveData(torch.utils.data.Dataset):
    def __init__(self,file_path,transform = None):
        self.data = pd.read_csv(file_path)
        self.len = len(self.data)
        
        self.label = np.asarray(self.data["label"])
        self.data = np.asarray(self.data.drop("label",axis = 1))
    def __getitem__(self,index):
        label = self.label[index]
        img = self.data[index].reshape(28,28).astype("uint8")
        img = Image.fromarray(img)
        img = img.convert("L")
        if(transform != None):
            img = transform(img)
        return img,label
    def __len__(self):
        return self.len
```

```python
# load data and check for cuda

transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,),(0.5,))])
trainSet = DriveData("./train.csv",transform)
testSet = DriveData("./test.csv",transform)

classSize = trainSet.label.max() + 1
trainOnGPU = torch.cuda.is_available()

if trainOnGPU:
    print("CUDA available... training on GPU")
else:
    print("CUDA not available... training on CPU")
```

```python
# split into training and validation
valSize = 0.2
numOfWorkers = 0
batchSize = 20

trainSize = len(trainSet)
indices = list(range(trainSize))
np.random.shuffle(indices)

split = int(np.floor(trainSize * valSize))
trainIdx,valIdx = indices[split:],indices[:split]

from torch.utils.data.sampler import SubsetRandomSampler

trainSampler = SubsetRandomSampler(trainIdx)
valSampler = SubsetRandomSampler(valIdx)

trainLoader = torch.utils.data.DataLoader(trainSet,batch_size = batchSize,num_workers = numOfWorkers, sampler = trainSampler)
valLoader = torch.utils.data.DataLoader(trainSet,batch_size = batchSize,num_workers = numOfWorkers, sampler = valSampler)
testLoader = torch.utils.data.DataLoader(testSet,batch_size = batchSize,num_workers = numOfWorkers)
```

```python
#visualize data

import matplotlib.pyplot as plt
%matplotlib inline

it = iter(trainLoader)
images,labels = it.next()
images = images.numpy()
fig = plt.figure(figsize=(25,5))

for idx in np.arange(20):
    ax = fig.add_subplot(2,20/2,idx+1,xticks=[],yticks=[])
    ax.imshow(np.squeeze(images[idx]),cmap="gray")
    
```

```python
from torch import nn,optim

class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
                                # sees 28x28x1
        self.conv1 = nn.Conv2d(1,16,3,padding=1)
                                # sees 14x14x16
        self.conv2 = nn.Conv2d(16,32,3,padding=1)
                                # sees 7x7x32
        self.conv3 = nn.Conv2d(32,64,3,padding=1)
                                # sees 3x3x64
        self.pool = nn.MaxPool2d(2,2)
        self.fc1 = nn.Linear(64*3*3,512)
        self.fc2 = nn.Linear(512,128)
        self.fc3 = nn.Linear(128,64)
        self.fc4 = nn.Linear(64,25)
        self.relu = nn.ReLU()
        self.softmax = nn.LogSoftmax(dim=1)
        self.dropout = nn.Dropout(0.2)
    def forward(self,x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = self.pool(self.relu(self.conv3(x)))
        
        # flatten the image
        x = x.view(-1,64*3*3)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.relu(self.fc3(x))
        x = self.dropout(x)
        x = self.fc4(x)
        x = self.softmax(x)
        return x
model = Classifier()
print(model)
```

```python
criterion = nn.NLLLoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
```

```python
epochs = 20

for epoch in range(epochs):
    
    valLoss = 0.0
    trainLoss = 0.0
    testLoss = 0.0
    accuracy = 0.0
    model.train()
    for data,label in trainLoader:
        if(trainOnGPU):
            data,label = data.cuda(),label.cuda()
        
        optimizer.zero_grad()
        output = model(data)
        loss =criterion(output,label)
        loss.backward()
        optimizer.step()
        trainLoss += loss.item()*data.size(0)
    with torch.no_grad():
        model.eval()
        
        for data,label in valLoader:
            if trainOnGPU:
                data,label = data.cuda(),label.cuda()
            output = model(data)
            loss = criterion(output,label)
            valLoss += loss.item() * data.size(0)
        for data,label in testLoader:
            model.eval()
            if trainOnGPU:
                data,label = data.cuda(),label.cuda()
            log_ps = model(data)
            testLoss += criterion(log_ps,label)
            ps = torch.exp(log_ps)
            top_p,top_c = ps.topk(1,dim=1)
            equals = top_c ==label.view(*top_c.shape)
            accuracy += torch.mean(equals.type(torch.FloatTensor))
    trainLoss = trainLoss / len(trainLoader.sampler)
    valLoss = valLoss / len(valLoader.sampler)
    accuracy = accuracy / len(testLoader) * 100.0
    print("Epoch {}\t Training loss: {:.6f}\t Validation loss: {:.6f}\t Test Accuracy: {:.2f}%".format(epoch,trainLoss
                                                                                                       ,valLoss,accuracy))
```

```python
model.eval()
it = iter(testLoader)
data,label = it.next()
output = model(data)
_,preds = torch.max(output,1)
fig = plt.figure(figsize=(25,4))
for idx in np.arange(20):
    ax = fig.add_subplot(2,20/2,idx+1,xticks=[],yticks=[])
    ax.imshow(np.squeeze(data[idx]),cmap="gray")
    ax.set_title("{} ({})".format(str(preds[idx].item()),str(label[idx].item())),
                color = ("green" if preds[idx] == label[idx] else "red"))
```
